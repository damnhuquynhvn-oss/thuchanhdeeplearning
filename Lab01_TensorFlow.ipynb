```python
def a_regular_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x

a_function_that_uses_a_graph = tf.function(
    a_regular_function
)

x1 = tf.constant([[1.0,2.0]])
y1 = tf.constant([[2.0],[3.0]])
b1 = tf.constant(4.0)

orig_value = a_regular_function(x1,y1,b1).numpy()

tf_function_value = a_function_that_uses_a_graph(
    x1,y1,b1
).numpy()

print(orig_value)
print(tf_function_value)
```

```python
def inner_function(x,y,b):
    x = tf.matmul(x,y)
    x = x + b
    return x

@tf.function
def outer_function(x):
    y = tf.constant([[2.0],[3.0]])
    b = tf.constant(4.0)

    return inner_function(x,y,b)

result = outer_function(
    tf.constant([[1.0,2.0]])
)

print(result.numpy())
```

```python
class SimpleModule(tf.Module):

    def __init__(self, name=None):
        super().__init__(name=name)

        self.a_variable = tf.Variable(5.0)

    def __call__(self, x):
        return self.a_variable * x


simple_module = SimpleModule()

print(simple_module(tf.constant(2.0)))

print("Variables:")
for var in simple_module.variables:
    print(var)
```

```python
class SimpleModule(tf.Module):

    def __init__(self, name=None):
        super().__init__(name=name)

        self.a_variable = tf.Variable(5.0)

    def __call__(self, x):
        return self.a_variable * x


simple_module = SimpleModule()

print(simple_module(tf.constant(2.0)))

print("Variables:")
for var in simple_module.variables:
    print(var)
```

```python
class Dense(tf.Module):

    def __init__(self, in_features, out_features, name=None):
        super().__init__(name=name)

        self.w = tf.Variable(
            tf.random.normal([in_features, out_features]),
            name='w'
        )

        self.b = tf.Variable(
            tf.zeros([out_features]),
            name='b'
        )

    def __call__(self, x):
        y = tf.matmul(x, self.w) + self.b
        return tf.nn.relu(y)


class SequentialModule(tf.Module):

    def __init__(self, name=None):
        super().__init__(name=name)

        self.dense_1 = Dense(3,3)
        self.dense_2 = Dense(3,2)

    def __call__(self, x):
        x = self.dense_1(x)
        return self.dense_2(x)


my_model = SequentialModule()

print(
    my_model(
        tf.constant([[2.0,2.0,2.0]])
    )
)

chk_path = "./my_checkpoint"

checkpoint = tf.train.Checkpoint(
    model=my_model
)

checkpoint.write(chk_path)

print("Saved checkpoint")
```

```python
new_model = SequentialModule()

new_checkpoint = tf.train.Checkpoint(
    model=new_model
)

new_checkpoint.restore("./my_checkpoint")

print(
    new_model(
        tf.constant([[2.0,2.0,2.0]])
    )
)
```
